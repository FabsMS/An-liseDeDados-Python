```python
df2018['data_inversa'] = pd.to_datetime(df2018['data_inversa'])
df2019['data_inversa'] = pd.to_datetime(df2019['data_inversa'])
df2020['data_inversa'] = pd.to_datetime(df2020['data_inversa'])
df2018.info()
df2019.info()
df2020.info()
```

<div class="alert alert-block alert-info">
Os passos feitos logo antes desse comentário foram necessários para que a coluna "data_inversa" mudassem de 'object' para 'datetime', e o que vem a ser realizado logo à seguir, possa ser feito com êxito
</div>

```python
agrupamento2018 = df2018.groupby(df2018['data_inversa'].dt.month)['feridos_graves'].sum()
agrupamento2018final = pd.DataFrame(agrupamento2018)
agrupamento2018final['ilesos'] = df2018.groupby(df2018['data_inversa'].dt.month)['ilesos'].sum()
totalFeridos = df2018.groupby(df2018['data_inversa'].dt.month)['feridos'].sum()
agrupamento2018final['total_pessoas'] = df2018.groupby(df2018['data_inversa'].dt.month)['pessoas'].sum()
agrupamento2018final['proporção_ilesos/total'] = agrupamento2018final['ilesos']/agrupamento2018final['total_pessoas']
agrupamento2018final['proporção_graves/total'] = agrupamento2018final['feridos_graves']/agrupamento2018final['total_pessoas']
agrupamento2018final
```

```python
agrupamento2019 = df2019.groupby(df2019['data_inversa'].dt.month)['feridos_graves'].sum()
agrupamento2019final = pd.DataFrame(agrupamento2019)
agrupamento2019final['ilesos'] = df2019.groupby(df2019['data_inversa'].dt.month)['ilesos'].sum()
totalFeridos = df2019.groupby(df2019['data_inversa'].dt.month)['feridos'].sum()
agrupamento2019final['total_pessoas'] = df2019.groupby(df2019['data_inversa'].dt.month)['pessoas'].sum()
agrupamento2019final['proporção_ilesos/total'] = agrupamento2019final['ilesos']/agrupamento2019final['total_pessoas']
agrupamento2019final['proporção_graves/total'] = agrupamento2019final['feridos_graves']/agrupamento2019final['total_pessoas']
agrupamento2019final
```

```python
df2020_2 = df2020
df2020_2['meses'] = df2020_2['data_inversa'].dt.month
pos_pandemia = (df2020_2['meses']>= 3).sum()
pos_pandemia
```

```python
df2020_3 = df2020
df2020_3['meses'] = df2020_3['data_inversa'].dt.month
pre_pandemia = (df2020_3['meses']< 3).sum()
pre_pandemia
```

```python
#Soma total de acidentes pré-pandemia
total_pre = len(df2018) + len(df2019)
total_pre = total_pre + pre_pandemia
total_pre 
```

<div class="alert alert-block alert-info">
Nos passos anteriores, foi separado, primeiro, a quantidade de acidentes ocorridos à partir de Março de 2020, momento em que o estado pandêmico foi estabelecido no Brasil<br>
Após isso, foram coletadas a quantidade de acidentes ocorridos em Janeiro e Fevereiro de 2020, meses em que ainda não havia pandemia no Brasil, e com isso, essa quantidade encontrada foi somada à quantidade total de acidentes nos anos anteriores
</div>

```python
#Proporções pedidas
QTD_MESES_PRE = 26 #Quantidade de meses no período pré-pandemia
QTD_MESES_POS = 10 #Quantidade de meses no período pós-pandemia
media_pos = "{:.2f}".format(pos_pandemia/10)
media_pre = "{:.2f}".format(total_pre/26)
print("média de acidentes, por mês, com a pandemia 'ativa':", media_pos)
print("média de acidentes, por mês, sem a pandemia 'ativa':", media_pre)
```

<div class="alert alert-block alert-info">
Nas divisões anteriores, temos a proporção, ou média, de acidentes que ocorreram, por mês, nos períodos pré e pós pandêmicos, os quais serão apresentados também no DataFrame à seguir.

```python
total_acidentes = [total_pre, pos_pandemia]
medias = [media_pre, media_pos]

dfMedias = pd.DataFrame({'total_acidentes':total_acidentes, 
                         'proporção':medias})
dfMedias
```
