    <p>Primeiramente, foi necessário converter as colunas 'stop_date' e 'stop_time' para datetime, após isso, com o uso da função 'groupby, a média entre "is_arrested", por hora, foi calculada</p>

```python
df_police['stop_time'] = pd.to_datetime(df_police['stop_time'])
df_police['stop_date'] = pd.to_datetime(df_police['stop_date'])
df_police.dtypes
```

```python
hourly_arrest_rate = pd.DataFrame(df_police.groupby(df_police['stop_time'].dt.hour)['is_arrested'].mean())
hourly_arrest_rate
```

<hr>

<div class="alert alert-block alert-info">
<h2>d)</h2>
    <p>Fazendo a média da coluna 'drugs_related_stop' em cada ano em quetsão, a partir do gruopby com a função 'mean', dados foram obtidos, e com esses dados, novamente usando a função 'plot' da biblioteca matplotlib, um gráfico de linhas foi criado</p>
</div>

```python
year_drugs = pd.DataFrame(df_police.groupby(df_police['stop_date'].dt.year)['drugs_related_stop'].mean())
year_drugs
```

```python
plt.plot(year_drugs)
plt.xlabel("Ano")
plt.ylabel("Taxa de Apreensão de Drogas")
plt.title("Taxa de Apreensão por Drogas por Ano")
plt.grid()
```

<hr>

<div class="alert alert-block alert-info">
<h2>e)</h2>
    <p>Após a obtenção das médias desejadas, foi feita a concateção dos dataframes desejados, sendo eles "annual_drug_rate" e "year_drugs".</p>
    <p>Com isso feito, foi feito um plot duplo, criando dois gráficos de linhas juntos, podendo, assim, analisar melhor os dados obtidos</p>
</div>

```python
annual_drug_rate = pd.DataFrame(df_police.groupby(df_police['stop_date'].dt.year)['search_conducted'].mean())
annual_drug_rate
```

```python
annual_search_drug = pd.concat([annual_drug_rate, year_drugs], ignore_index=False, axis=1)
annual_search_drug
```

```python
plt.plot(year_drugs, label='Apreensão por drogas')
plt.plot(annual_drug_rate, label='Taxa de paradas')
plt.xlabel("Ano")
plt.ylabel("Taxa")
plt.title("Taxa Anual de Drogas e Taxa Anual de Apreensões")
plt.legend()
plt.grid()
```
